```python
import matplotlib.pyplot as plt
import seaborn as sns
from statannot import add_stat_annotation
import pandas as pd
import numpy as np
```

```python
results_rf=pd.read_csv('results_rf_with_prc.csv',index_col=0)
results_xgb=pd.read_csv('results_xgb_with_prc.csv',index_col=0)
results_DT=pd.read_csv('results_rf_with_prc.csv',index_col=0)
```

```python
results_rf_prc=results_rf.loc[results_rf['Metric']=='prc'].reset_index(drop=True)
results_rf_prc['Classifier']=np.repeat('Random Forest',len(results_rf_prc))

results_xgb_prc=results_xgb.loc[results_xgb['Metric']=='prc'].reset_index(drop=True)
results_xgb_prc['Classifier']=np.repeat('XGBoost',len(results_xgb_prc))

results_dt_prc=results_DT.loc[results_DT['Metric']=='prc'].reset_index(drop=True)
results_dt_prc['Classifier']=np.repeat('Decision Tree',len(results_dt_prc))

results_prc=pd.concat([results_rf_prc,results_xgb_prc,results_dt_prc])
#results_acc=pd.concat([results_rf_acc,results_xgb_acc])

results_prc=results_prc.rename(columns={'Score':'AU prc'})
```

```python
plt.figure(dpi=300)
x = "Classifier"
y = "AU prc"
order = ['Random Forest', 'XGBoost', 'Decision Tree']
ax = sns.boxplot(data=results_prc, x=x, y=y, order=order,color='cyan')
#add_stat_annotation(ax, data=results_acc, x=x, y=y, order=order,
 #                   box_pairs=[("Random Forest", "XGBoost"), ("Random Forest", "Decision Tree"), ("XGBoost", "Decision Tree")],
  #                  test='Mann-Whitney', text_format='star', loc='outside', verbose=2)
#add_stat_annotation(ax, data=results_acc, x=x, y=y, order=order,
                    #box_pairs=[("Random Forest", "XGBoost")], test='Mann-Whitney', text_format='star', loc='outside', verbose=2)#("Random Forest", "Decision Tree"), ("XGBoost", "Decision Tree")],
                    #test='Mann-Whitney', text_format='star', loc='outside', verbose=2)
add_stat_annotation(ax, data=results_prc, x=x, y=y, order=order,
                    box_pairs=[("Random Forest", "XGBoost"), ("Random Forest", "Decision Tree"), ("XGBoost", "Decision Tree")],
                    test='Mann-Whitney', text_format='star', loc='outside', verbose=2)

ax.grid(False)
```

```python
prob_dt=pd.read_csv('prob_dt.csv',index_col=0)
prob_rf=pd.read_csv('prob_rf.csv',index_col=0)[['Probability','Target']]
prob_xgb=pd.read_csv('prob_xgb.csv',index_col=0)[['Probability','Target']]
prob_xgb
```

```python
#ROC Curve
from sklearn.metrics import roc_curve

y_pred_prob1 = prob_dt['Probability']
fpr1 , tpr1, thresholds1 = roc_curve(prob_dt['Target'], y_pred_prob1,pos_label='ASD')

y_pred_prob2 = prob_xgb['Probability']
fpr2 , tpr2, thresholds2 = roc_curve(prob_xgb['Target'], y_pred_prob2,pos_label='ASD')


y_pred_prob3 = prob_rf['Probability']
fpr3 , tpr3, thresholds3 = roc_curve(prob_rf['Target'], y_pred_prob3,pos_label='ASD')


fig, ax = plt.subplots(dpi=300)
plt.plot([0,1], [0,1], linestyle='--', lw=2,color='#356c6e', label='Chance',alpha=0.8)
plt.plot(fpr1, tpr1, color='#f4cda9',
         label=r'Mean ROC DT (AUC = %0.3f $\pm$ %0.3f)' % (0.837, 0.022), lw=2, alpha=.8)
#plt.plot(fpr1, tpr1, label= "DT")
plt.plot(fpr2, tpr2, color='#6c1602',
         label=r'Mean ROC XGB (AUC = %0.3f $\pm$ %0.3f)' % (0.966, 0.004), lw=2, alpha=.8)
#plt.plot(fpr2, tpr2, label= "XGB")
plt.plot(fpr3, tpr3, color='#fd722e',
         label=r'Mean ROC RF (AUC = %0.3f $\pm$ %0.3f)' % (0.965, 0.003), lw=2, alpha=.8)
#plt.plot(fpr3, tpr3, label= "RF")

plt.legend()
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.01])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
#plt.title('Receiver Operating Characteristic')
plt.show()
```
